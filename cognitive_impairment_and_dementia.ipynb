```python
import pandas as pd
import re
```

```python
import requests

# Let's check out this website
url = "https://en.wikipedia.org/wiki/Cognitive_impairment"
response = requests.get(url)
response
```

```python
response.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
text = ""
```

```python
for paragraph in soup.find_all('p'):
    text+= paragraph.text
```

```python
text = re.sub(r'\[[0-9]*\]', ' ', text)
text = re.sub(r'\b(t)\b', r' \1 ', text)
text = text.lower()
```

```python
with open('cognitive_impairment.txt', 'w', encoding='utf-8') as file:
    file.write(text)

print("Cleaned text saved to cognitive_impairment.txt")
```

```python
url_1 = "https://en.wikipedia.org/wiki/Dementia"
response_1 = requests.get(url_1)
response_1
```

```python
response_1.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response_1.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
text = ""
```

```python
# Extract paragraphs and add them to text
for paragraph in soup.find_all('p'):
    text += paragraph.text + "\n\n"

# Extract headings (including those within <span> elements) and add them to text
for heading in soup.find_all(['h2']):
    text += heading.text + "\n"

# Remove reference numbers like [1], [2], etc.
text = re.sub(r'\[[0-9]*\]', ' ', text)

# Save the cleaned text to a text file
with open('dementia_content.txt', 'w', encoding='utf-8') as file:
    file.write(text)

print("Cleaned text saved to dementia_content.txt")
```
